````python
#1) Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), 
# e que devolve os dados da consulta no formato DataFrame.

url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl' #inserindo a url com todos os estados
r = requests.get(url)
````

````python
data_json = r.json()
df_brasil = pd.DataFrame(data_json['items'])
df_brasil
````

````python
# Criando a função

def consulta_dados(uf, tipo_interessado):
    # Filtra os dados com base no UF e tipo de interessado
    dados_filtrados = df_brasil[(df_brasil['uf'] == uf) & (df_brasil['tipo_interessado'] == tipo_interessado)]
    return dados_filtrados
````

````python
uf_consulta = 'SP'
tipo_interessado_consulta = 'Estado'
resultado_consulta = consulta_dados(uf_consulta, tipo_interessado_consulta)
resultado_consulta
````
